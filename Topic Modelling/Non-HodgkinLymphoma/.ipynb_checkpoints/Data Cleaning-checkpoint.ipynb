```python
import re
```

```python
def find_lymphoma(x):
    if re.search('lymphoma',x):
        return True
    else:
        return False
```

```python
df['is_lymphoma'] = df['Full Text'].apply(find_lymphoma)
```

```python
def find_hodgkin(x):
    x = x.lower()
    if re.search(r'hodgkin',x):
        return True
    else:
        return False
```

```python
df['is_hodgkin'] = df['Full Text'].apply(find_hodgkin)
```

```python
df_hodgkin = df[df['is_hodgkin']].copy()
df_hodgkin.shape
```

```python
df.drop('is_lymphoma' , inplace = True , axis = 1)
```
